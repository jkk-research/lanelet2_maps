```python
import mat4py
import numpy as np
import pandas as pd
```

```python
# read matlab file
data = mat4py.loadmat('/home/matyko/Downloads/lanes_ZZ.mat')
```

```python
print(data["lanes"].keys())
```

```python
path = np.array(data["lanes"]["path"][0][0])
path_2 = np.array(data["lanes"]["path"][0][1])
path_3 = np.array(data["lanes"]["path"][0][2])
```

```python
#plot lane1 using plotly
import plotly.graph_objects as go

fig = go.Figure()
# plot with the same scale for x and y
fig.update_xaxes(scaleanchor = "y", scaleratio = 1)

fig.add_trace(go.Scatter(x=path[:,0], y=path[:,1], mode='lines', name='lane1'))
fig.add_trace(go.Scatter(x=path_2[:,0], y=path_2[:,1], mode='lines', name='lane2'))
fig.add_trace(go.Scatter(x=path_3[:,0], y=path_3[:,1], mode='lines', name='lane3'))

```
