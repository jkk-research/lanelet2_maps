```python
#!/usr/bin/env python
# coding: utf-8
import sys
import os
import xml.etree.ElementTree as ET
from pyproj import Proj, transform
```

```python
def wgs84_to_utm(lat, lon):
    # Define the UTM zone based on longitude
    utm_zone = int((lon + 180) / 6) + 1

    # Define projections: WGS84 and UTM
    wgs84_proj = Proj(proj="latlong", datum="WGS84")
    utm_proj = Proj(proj="utm", zone=utm_zone, ellps="WGS84", datum="WGS84", south=False)

    # Transform the coordinates
    easting, northing = transform(wgs84_proj, utm_proj, lon, lat)
    return easting, northing, utm_zone
```

```python
if len(sys.argv) > 1:

    filename = sys.argv[1]

    tree = ET.parse("/home/matyko/Downloads/lightweight_ll.osm")
```

```python
root = tree.getroot()
for child in root:
    if child.tag == 'node':
        if 'lat' in child.attrib and 'lon' in child.attrib:
            lat = float(child.attrib['lat'])
            lon = float(child.attrib['lon'])
            easting, northing, utm_zone = wgs84_to_utm(lat, lon)
            print(easting, northing, utm_zone)
            child.attrib['lat'] = ''
            child.attrib['lon'] = ''

        for grandchild in child:
            if grandchild.attrib['k'] == 'local_x':
                grandchild.attrib['v'] = str(float(easting) - 639770.000)
            if grandchild.attrib['k'] == 'local_y':
                grandchild.attrib['v'] = str(float(northing) - 5195040.000 )

        if 'local_x' in child.attrib and 'local_y' in child.attrib:
            
            child.attrib['local_x']['v'] = str(easting)
            child.attrib['local_y']['v'] = str(northing)
```
