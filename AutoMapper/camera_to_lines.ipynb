```python
import pandas as pd
import numpy as np
import mat4py
from scipy.io import loadmat
import glob
import os
```

```python
current_dir = os.getcwd()
folder_path = current_dir + "/raw_lines"  # Change this to your folder path
mat_files = glob.glob(f"{folder_path}/*.mat")

print(mat_files)
```

```python
measurements = []
```

```python
for mat_file in mat_files:
    data = loadmat(mat_file)
    measurements.append(data)
```

```python
#convert measurements to pandas dataframe
measurements = pd.DataFrame(measurements)
```
